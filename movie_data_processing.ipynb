```python
import pandas as pd
import numpy as np
```

```python
# Assuming the file is in the same directory as your Python script
file_path = "imdb_data/title.basics.tsv"

# Load the TSV file into a pandas DataFrame
movie_title_df = pd.read_csv(file_path, sep='\t')

movie_title_df['startYear'] = pd.to_numeric(movie_title_df['startYear'], errors='coerce')
movie_title_df['runtimeMinutes'] = pd.to_numeric(movie_title_df['runtimeMinutes'], errors='coerce')
# Display the first few rows of the DataFrame to verify it loaded correctly
movie_title_df.info()
```

```python
movie_title_df = movie_title_df
movie_title_df[(movie_title_df['startYear'] >= 1970) & (movie_title_df['runtimeMinutes'] >= 30)]
movie_title_df.head()
```

```python
movie_title_df.info()
```

```python
filtered_movies_df = pd.read_csv('clean_data/movie_ratings_clean.csv')
filtered_movies_df.info()
filtered_movies_df.head()
```

```python
filtered_movie_title_df = movie_title_df[movie_title_df['tconst'].isin(filtered_movies_df['tconst'])]
filtered_movie_title_df.info()
```

```python
filtered_movie_title_df = filtered_movie_title_df.dropna()
filtered_movie_title_df.info()
```

```python
filtered_movie_title_df.head()
```

```python
filtered_movies_df = filtered_movies_df[filtered_movies_df['tconst'].isin(filtered_movie_title_df['tconst'])]
filtered_movies_df.info()
```

```python
filtered_movies_df.to_csv('clean_data/movie_ratings_clean.csv', index=False)
filtered_movie_title_df.to_csv('clean_data/movie_title_clean.csv', index=False)
```

```python
name_df = pd.read_csv('imdb_data/name.basics.tsv', sep='\t')
movie_crew_df = pd.read_csv('imdb_data/title.crew.tsv', sep='\t')
principals_df = pd.read_csv('imdb_data/title.principals.tsv', sep='\t')
```

```python
name_df.head()
```

```python
movie_crew_df.head()

filtered_movie_crew = movie_crew_df[movie_crew_df['tconst'].isin((filtered_movie_title_df['tconst']))]
filtered_movie_crew.info()
filtered_movie_crew.head()
```

```python
import pandas as pd

# Assuming filtered_movie_crew is already defined and includes 'directors' and 'writers' columns

# Splitting and expanding the directors list, then getting unique values
directors_list = filtered_movie_crew['directors'].str.split(',').explode().unique()

# Splitting and expanding the writers list, then getting unique values
writers_list = filtered_movie_crew['writers'].str.split(',').explode().unique()

# Converting the lists to sets
directors_set = set(directors_list)
writers_set = set(writers_list)

# Combining both sets to get a set of all unique directors and writers
directors_writers_set = directors_set | writers_set  # Using the union operator
len(directors_writers_set)
```

```python
directors_writers_list = list(directors_writers_set)  # Convert set to list

# Create a DataFrame from this list
directors_writers_df = pd.DataFrame(directors_writers_list, columns=['nconst'])

directors_writers_df.head()
```

```python
filtered_principals_df = principals_df[principals_df['tconst'].isin(filtered_movie_title_df['tconst'])]

filtered_principals_df = filtered_principals_df.drop_duplicates(subset=['tconst', 'nconst'])
filtered_principals_df.info()
filtered_principals_df.head()
```

```python
nconst_df = filtered_principals_df[['nconst']].drop_duplicates()

combined_nconst_df = combined_df = pd.concat([directors_writers_df, nconst_df], ignore_index=True).drop_duplicates()
combined_nconst_df.info()
```

```python
filtered_names_df = name_df[name_df['nconst'].isin(combined_nconst_df['nconst'])]
filtered_movies_df.info()
filtered_names_df.head()
```

```python
filtered_names_df.to_csv('clean_data/names_clean.csv', index=False)
filtered_principals_df.to_csv('clean_data/principals_clean.csv', index=False)
filtered_movie_crew.to_csv('clean_data/crew_clean.csv', index=False)
```
